```python
import os
import pandas as pd
```

```python
for disease in ['BLOOD PRESSURE', 'FEVER', 'FIRST AID', 'GAS TROUBLE', 'SOAPS', 'SUGAR', 'THYROID']:
    for years in ['2018 - 2019', '2019 - 2020']:
        path = 'C:/Users/User/Desktop/medical project/DEMAND ANALYSIS2/{}/{}'.format(disease, years)
        files = os.listdir(path)
        
        #extracting only files contaning .xls format
        data = []
        for f in files:
            if f[-3:] == 'xls':
                data.append(f)
                for name in data:
                    df = pd.read_excel('C:/Users/User/Desktop/medical project/DEMAND ANALYSIS2/{}/{}/{}'
                       .format(disease, years, name), encoding = 'ISO-8859-1')
                    df = df.dropna(axis = 0, how = 'all')

                    df.columns = ['Doc. Date','a','Ref. No','Party','NaN','CITY','Batch','ORDER_FREE','ORDER_FREE_QUANTITY','SOLD_QUANTITY','SOLD_FREE','Bal. Qty']
                    df = df.fillna(0)

                    df2  = df.iloc[4]['a']
                    df4 = df2.replace('/', ' ')
                    
                    df3 = df.iloc[2]['SOLD_FREE']
                    df3 = df3.split(':')[1]
                    df3 = df3.split('/')
                    df3 = df3[-1]
                    
                    df['Name of Tablet'] = str(df4)

                    df = df.iloc[6:-2]
                    
                    df['Year'] = df['Doc. Date'].apply(lambda x: str(x).split('-')[0])
                    df['Month'] = df['Doc. Date'].apply(lambda x: str(x).split('-')[1])
                    df['Day2'] = df['Doc. Date'].apply(lambda x: str(x).split('-')[2])
                    df['Day'] = df['Day2'].apply(lambda x: str(x).split()[0])

                    cols = ['Month', 'Day','Year']
                    df['Date'] = df[cols].apply(lambda row:'/'.join(row.values.astype(str)), axis=1)

                    df = df.drop(columns=['Batch','Bal. Qty','Doc. Date', 'Day2','a','Ref. No','Party','NaN','Day2'])
                    df = df.reset_index(drop = True)
                    

                    df.to_csv('C:/Users/User/Desktop/medical project/DEMAND ANALYSIS2/files4/{}.csv'.format( df4 + df3 ), index= False)
```

```python
files = [file for file in os.listdir('C:/Users/User/Desktop/medical project/DEMAND ANALYSIS2/files4')]

data = pd.DataFrame()

for file in files:
    df2 = pd.read_csv('C:/Users/User/Desktop/medical project/DEMAND ANALYSIS2/files4/' + file)
    data = pd.concat([data, df2])
data = data.reset_index(drop=True)

data
```
